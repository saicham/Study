```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans, DBSCAN
from sklearn.datasets import make_blobs

#  
X, y = make_blobs(n_samples=300, centers=4, cluster_std=0.60, random_state=0)

# K-Means 
kmeans = KMeans(n_clusters=4)
kmeans.fit(X)
kmeans_labels = kmeans.predict(X)

# DBSCAN 
dbscan = DBSCAN(eps=0.5, min_samples=5)
dbscan.fit(X)
dbscan_labels = dbscan.labels_

# 
fig, axs = plt.subplots(ncols=2, figsize=(10, 5))
sns.scatterplot(x=X[:, 0], y=X[:, 1], hue=y, ax=axs[0])
axs[0].set_title('Original Data')
sns.scatterplot(x=X[:, 0], y=X[:, 1], hue=kmeans_labels, ax=axs[1])
axs[1].set_title('K-Means Clustering')
plt.show()

fig, axs = plt.subplots(ncols=2, figsize=(10, 5))
sns.scatterplot(x=X[:, 0], y=X[:, 1], hue=y, ax=axs[0])
axs[0].set_title('Original Data')
sns.scatterplot(x=X[:, 0], y=X[:, 1], hue=dbscan_labels, ax=axs[1])
axs[1].set_title('DBSCAN Clustering')
plt.show()
```

```python
from sklearn.cluster import KMeans
import pandas as pd
import matplotlib.pyplot as plt

#  
df = pd.DataFrame({'B': [1, 2, 3, '', '', ''], 
                   'C': ['-', '-', '-', '1', '2', '3']})

#  
X = df[df['B'] != '']['B'].values.reshape(-1, 1)

# KMeans 
kmeans = KMeans(n_clusters=3, random_state=0).fit(X)

#  
plt.scatter(X, [0]*len(X), c=kmeans.labels_)
plt.scatter(kmeans.cluster_centers_, [0]*len(kmeans.cluster_centers_), c='r', marker='x')
plt.show()
```

```python
from sklearn.cluster import KMeans
from gensim.models import Word2Vec
import numpy as np

#   
sentences = [[''], ['1'], ['2'], ['3']]
model = Word2Vec(sentences, min_count=1)

#   
X = np.zeros((4, 100))
X[0] = model.wv['']
X[1] = model.wv['1']
X[2] = model.wv['2']
X[3] = model.wv['3']

# kmeans 
kmeans = KMeans(n_clusters=2, random_state=0).fit(X)
print(kmeans.labels_)

```
