### xor 연산자


```python
# XOR연산결과를 반환하는 함수 생성
def xor(x,y):
    return bool((x and not y) or (not x and y))
```

```python
# 값을 입력받아 xor연산결과를 화면에 출력
a = int(input("a를 입력하세요: "))
b = int(input("b를 입력하세요: "))

print(xor(a,b))
```
